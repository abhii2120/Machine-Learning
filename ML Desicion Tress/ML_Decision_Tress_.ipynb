##Q1. What is a Decision Tree? Explain its structure
###Answer:

A Decision Tree is a supervised machine learning algorithm used for solving both classification and regression problems. It works by splitting the dataset into smaller subsets based on certain conditions and represents these decisions in a tree-like structure.

The algorithm selects the best feature at each step to divide the data, such that the resulting groups are as pure as possible (i.e., similar outcomes are grouped together). This process continues recursively until a stopping condition is met.

Decision Trees are widely used because they are simple, easy to understand, and do not require complex mathematical computations.

##Structure of Decision Tree

A Decision Tree consists of the following main components:

###1. Root Node
The root node is the top-most node in the tree
It represents the entire dataset
The first split or decision is made here
It is selected based on the most important feature

#### Example:
“Age > 30?”

This node divides the dataset into two parts based on the condition.

###2. Decision Nodes
These nodes represent further conditions or rules
Each node splits the data based on a feature
Multiple decision nodes can exist in a tree

####Example:
“Income > 50,000?”
“Education Level = Graduate?”

###These nodes continue dividing the dataset into smaller subsets.

###3. Leaf Nodes
These are the final nodes of the tree
No further splitting occurs here
They represent the final prediction or outcome

####Example:
“Yes (Customer will subscribe)”
“No (Customer will not subscribe)”

###4. Branches
Branches represent the flow from one node to another
Each branch corresponds to a decision rule

####Example:
If Age > 30 → go to next condition

###How Decision Tree Works
Start with the entire dataset
Select the best feature to split the data
Divide the dataset into subsets
Repeat the process for each subset
Stop when:
Data is pure OR
Maximum depth is reached
###Real-Life Example

Consider a bank deciding whether a customer will subscribe to a fixed deposit scheme.

Decision Tree may look like:


Root: Age > 30

Yes → Income > 50K

Yes → Subscribe (Yes)

No → Not Subscribe (No)

No → Job = Student

Yes → Subscribe (Yes)

No → Not Subscribe (No)


This shows how decisions are made step-by-step logically.

###Advantages of Decision Tree
Easy to understand and interpret
Works with both numerical and categorical data
Requires little data preprocessing
Can handle non-linear relationships
###Disadvantages
Can overfit the data if not controlled
Sensitive to small changes in data
May become complex with large datasets
###Conclusion

A Decision Tree is a powerful and intuitive machine learning algorithm that mimics human decision-making. Its tree-like structure helps in breaking down complex problems into simple decisions, making it highly useful for real-world applications like banking, marketing, and healthcare.

###Q2. Differentiate between Gini Impurity and Entropy. Which one is used by default in Scikit-learn and why?
###Answer:

Gini Impurity and Entropy are criteria used in Decision Trees to measure how well a feature splits the dataset. Both aim to reduce impurity, meaning they try to create groups where data points belong to a single class as much as possible.

Gini Impurity

Gini Impurity measures the probability that a randomly selected data point would be incorrectly classified if it was randomly labeled according to the class distribution in that node.

Formula:
Gini = 1 − Σ (pᵢ)²

Where pᵢ is the probability of each class.

If all data points belong to one class, Gini = 0 (perfect purity)
If data is equally distributed, Gini is higher (more impurity)

It focuses on minimizing misclassification.

Entropy

Entropy measures the amount of uncertainty or randomness in the data. It is based on the concept of information theory.

Formula:
Entropy = − Σ pᵢ log₂(pᵢ)

If all data points belong to one class, Entropy = 0 (no uncertainty)
If data is evenly distributed, Entropy is maximum (high uncertainty)

It focuses on reducing randomness and maximizing information gain.
###Difference between Gini Impurity and Entropy

####Definition

Gini Impurity: Measures probability of wrong classification

Entropy: Measures uncertainty or randomness

####Formula

Gini = 1 − Σ (pᵢ)²

Entropy = − Σ pᵢ log₂(pᵢ)

####Calculation

Gini: Simple and fast

Entropy: Complex due to logarithm

####Speed

Gini: Faster

Entropy: Slower

####Usage

Gini: Commonly used

Entropy: Used when information gain needed

###Which one is used by default in Scikit-learn and why?

####Scikit-learn uses Gini Impurity as the default criterion in Decision Tree algorithms.

Reason:

It is computationally faster than Entropy
It does not involve logarithmic calculations
It provides similar performance compared to Entropy
It is efficient for large datasets
###Conclusion

Both Gini Impurity and Entropy are used to find the best split in a Decision Tree. However, Gini is preferred in Scikit-learn due to its simplicity and faster computation while still giving accurate results.

##Q3. What is Overfitting in Decision Trees? How can we detect it using training and testing accuracy?
###Answer:

Overfitting is a condition in which a Decision Tree model learns the training data too closely, including noise, outliers, and unnecessary details. Instead of capturing the general pattern of the data, the model memorizes the training dataset. As a result, it performs extremely well on training data but fails to perform well on new or unseen data.

In Decision Trees, overfitting commonly occurs when the tree grows too deep and creates many branches. Each branch tries to perfectly classify the training data, which makes the model very complex and less generalized.

###Why Overfitting Happens in Decision Trees
Decision Trees do not have built-in restrictions on depth by default
The model keeps splitting until all data points are perfectly classified
It captures noise along with actual patterns
Complex trees become highly sensitive to small changes in data
Detection using Training and Testing Accuracy

Overfitting can be detected by comparing the model’s performance on training data and testing data:

Training Accuracy: Measures how well the model performs on the data it was trained on
Testing Accuracy: Measures how well the model performs on unseen data
Condition for Overfitting:
Training accuracy is very high
Testing accuracy is significantly lower

This indicates that the model has memorized the training data but cannot generalize to new data.

###Example
Training Accuracy = 99%

Testing Accuracy = 68%

This large difference shows that the model is overfitting.

###Key Observation
Small difference → Good model
Large difference → Overfitting
###Conclusion

Overfitting in Decision Trees occurs when the model becomes too complex and captures noise instead of meaningful patterns. It can be easily detected by comparing training and testing accuracy. A large gap between the two indicates poor generalization and overfitting.

##Q4. Explain Pruning in Decision Trees. What is the difference between Pre-pruning and Post-pruning?
###Answer

Pruning in Decision Trees is a technique used to reduce the complexity of the model by removing unnecessary branches. It helps in improving the model’s performance on unseen data and prevents overfitting.

When a decision tree grows too deep, it starts capturing noise in the data instead of actual patterns. Pruning simplifies the tree by cutting off such branches, making the model more general and efficient.

##Types of Pruning
###1. Pre-pruning (Early Stopping)

Pre-pruning is applied during the tree-building process. It stops the growth of the tree before it becomes too complex.

###Methods used in pre-pruning:


Setting maximum depth (max_depth)

Minimum samples required to split (min_samples_split)

Minimum samples in leaf nodes (min_samples_leaf)

Working:

The tree stops splitting further when a predefined condition is met.

###2. Post-pruning (Backward Pruning)

Post-pruning is applied after the tree is fully grown. In this method, the

tree is first built completely and then unnecessary branches are removed.

Working:

Build full tree

Evaluate performance

Remove branches that do not improve accuracy
###Difference between Pre-pruning and Post-pruning
####Stage of Application

Pre-pruning: Applied during tree construction

Post-pruning: Applied after tree is built
###Approach
Pre-pruning: Stops tree growth early

Post-pruning: Removes branches after full growth

Risk

Pre-pruning: May stop too early (underfitting)

Post-pruning: More reliable, reduces overfitting effectively

Complexity

Pre-pruning: Simple and fast

Post-pruning: More computationally expensive
###Conclusion

Pruning is an important technique to control the complexity of Decision Trees. Pre-pruning stops the tree from growing too large, while post-pruning removes unnecessary branches after full growth. Both methods help in improving model generalization and reducing overfitting.

Feature Importance refers to a technique used in machine learning to identify which features (variables) in a dataset have the most influence on the model’s predictions. In Decision Trees, feature importance is calculated based on how much each feature contributes to reducing impurity (such as Gini or Entropy) during the splitting process.

Features that are used more frequently and result in better splits are considered more important.

###How Feature Importance Works

The total reduction in impurity contributed by each feature is calculated
Features with higher contribution are ranked as more important

###Explanation

The dataset is then split into training and testing sets using an 80-20 ratio. The training set is used to train the model, and the testing set is used to evaluate how well the model performs on unseen data.

###Interpretation

##Q10. Model Building
###Question Requirements
Train a Decision Tree Classifier

Use:

criterion = 'gini'

max_depth = 5

Interpret: What does max_depth control?

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Create Decision Tree model
model = DecisionTreeClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(15,8))

plot_tree(model, filled=True)

plt.title("Decision Tree (max_depth=5)")

plt.show()
```

###Explanation

In this step, a Decision Tree Classifier is trained using the training dataset. The parameter criterion='gini' is used to measure the quality of splits, which helps in selecting the best feature at each node.

The parameter max_depth=5 limits the maximum depth of the tree. This means the tree can only grow up to 5 levels, which helps in controlling the complexity of the model.

The model is trained using X_train and y_train, and predictions are made on the test dataset.

###Interpretation

The Decision Tree model was successfully trained using the Gini criterion. The tree splits the data based on important features to make predictions.

###What does max_depth control?

The max_depth parameter controls the maximum number of levels in the tree. A smaller depth reduces model complexity and helps prevent overfitting, while a larger depth makes the model more complex and may lead to overfitting.

###Conclusion

The Decision Tree model was successfully built using the specified parameters. Controlling the depth of the tree helps in balancing model accuracy and complexity.
